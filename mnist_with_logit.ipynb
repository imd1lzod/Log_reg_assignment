```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
X, y = df.drop(columns='23'), df['23']
```

```python
df_test = pd.read_csv("./data/letters-test.csv")
df_test
```

```python
df_test = df_test.loc[df_test['1'] < 5].reset_index(drop=True)
```

```python
X_test, y_test = df_test.drop(columns=['1']), df_test['1']
```

```python
X_test
```

```python
lr = LogisticRegression()
lr.fit(X.values, y.values)
# 
```

```python
preds = lr.predict(X_test.values)
```

```python
print(classification_report(y_test, preds))
```
